```r
# customer Gender Visualization.
```

```r
ggplot(df,aes(x=Gender, fill= Gender))+
geom_bar(color=rainbow(2))+xlab('Female/Male')
```

```r
df1=table(df$Gender)
df1
per=df1/sum(df1)*100
per
pie(x=per,labeles=paste(df$Gender," ",per," %"), col=rainbow(2),main="Pie chart")
```

```r
n_bins=10
ggplot(df, aes(x=Age))+geom_histogram(bins=n_bins, color='green',fill="Dark Green")
xlab('Age')
labs(title="histogram")
```

```r
ggplot(df, aes(x=Age))+ geom_boxplot()
```
